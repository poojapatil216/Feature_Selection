```python
import pandas as pd
import numpy as np
from scipy import stats 
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
df = pd.read_csv('test_data.csv')
feature_dict = {i:label for i,label in zip(
                range(21),
                  ('load_type',
                   'truck_type',
                   'route_type',
                   'rainfall',
                   'temperature',
                   'temp_type',
                   'tyre1_press',
                   'tyre1_press_stat',
                   'tyre2_press',
                   'tyre2_press_stat',
                   'tyre3_press',
                   'tyre3_press_stat',
                   'tyre4_press',
                   'tyre4_press_stat',
                   'tyre5_press',
                   'tyre5_press_stat',
                   'tyre6_press',
                   'tyre6_press_stat',
                   'avg_speed',
                   'time',
                   'fuel',
                   ))}

df.columns = [l for i,l in sorted(feature_dict.items())] + [
                   'tyre1_condition',
                   'tyre2_condition',
                   'tyre3_condition',
                   'tyre4_condition',
                   'tyre5_condition',
                   'tyre6_condition',
                   'efficiency',      ]
df.dropna(how="all", inplace=True) # to drop the empty line at file-end

df.tail()
```

```python
final_df1=pd.DataFrame()
final_df1['Feature']=['Temperature',
                     'Tyre 1 Pressure',
                     'Tyre 2 Pressure',
                     'Tyre 3 Pressure',
                     'Tyre 4 Pressure',
                     'Tyre 5 Pressure',
                     'Tyre 6 Pressure',
                     'Average Speed',
                     'Time',
                     'Fuel',]
scores=[]
pvalues=[]
```

```python

print('Temperature')
y=df['efficiency'].values.reshape(-1,1)
X=df['temperature'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)

scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 1 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre1_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 2 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre2_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 3 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre3_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 4 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre4_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 5 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre5_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Tyre 6 Pressure')
y=df['efficiency'].values.reshape(-1,1)
X=df['tyre6_press'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Average')
y=df['efficiency'].values.reshape(-1,1)
X=df['avg_speed'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Time')
y=df['efficiency'].values.reshape(-1,1)
X=df['time'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))

print('Fuel')
y=df['efficiency'].values.reshape(-1,1)
X=df['fuel'].values.reshape(-1,1)
X= StandardScaler().fit_transform(X)
y= StandardScaler().fit_transform(y)

fscores,p_values=f_regression(X,y)
scores.append(float(fscores))
pvalues.append(float(p_values))
```

```python
final_df1['F_Scores'] = scores
final_df1['P_Values']=pvalues
```

```python
y_new=pd.DataFrame()
X = df['efficiency'].values
X= StandardScaler().fit_transform(X)

y = df['load_type'].values
y_new['load_type']=y
cleanup_nums = {"load_type":     {"Heavy Load": 1, "Moderate": 0, "Full Load": 0.5}}
y_new.replace(cleanup_nums, inplace=True)

y = df['temp_type'].values
y_new['temp_type']=y
cleanup_nums = { "temp_type": {"Low": 0, "Normal": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre1_press_stat'].values
y_new['tyre1_press_stat']=y
cleanup_nums = { "tyre1_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre2_press_stat'].values
y_new['tyre2_press_stat']=y
cleanup_nums = { "tyre2_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre3_press_stat'].values
y_new['tyre3_press_stat']=y
cleanup_nums = { "tyre3_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre4_press_stat'].values
y_new['tyre4_press_stat']=y
cleanup_nums = { "tyre4_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre5_press_stat'].values
y_new['tyre5_press_stat']=y
cleanup_nums = { "tyre5_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)


y = df['tyre6_press_stat'].values
y_new['tyre6_press_stat']=y
cleanup_nums = { "tyre6_press_stat": {"Low": 0, "Correct": 0.5, "High": 1}}
y_new.replace(cleanup_nums, inplace=True)
```

```python
scores,pvalues=f_regression(y_new,X)

Final_features=pd.DataFrame()
Final_features['Feature']=['Load Type',
                   'Temp Type',
                   'Tyre1 Press Stat',
                   'Tyre2 Press Stat',
                   'Tyre3 Press Stat',
                   'Tyre4 Press Stat',
                   'Tyre5 Press Stat',
                   'Tyre6 Press Stat',
                   ]
Final_features['F_Scores'] = scores
Final_features['P_Values'] = pvalues
```
